```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import warnings
warnings.filterwarnings("ignore")

from math import pi
```

```python
agg_data=pd.read_csv('C:/Users/sun/Desktop/Univ/4_1/MediaProject/Project2/agg_match_stats_4.csv')
```

```python
kill_MIRAMAR=pd.read_csv('C:/Users/sun/Desktop/Univ/4_1/MediaProject/Project2/data_Kill_MIRAMAR.csv')
kill_ERANGEL=pd.read_csv('C:/Users/sun/Desktop/Univ/4_1/MediaProject/Project2/data_Kill_ERANGEL.csv')
```

```python
#kill데이터와 agg데이터 join하기(agg데이터에는 map 데이터 없음)
join_MIRAMAR=pd.merge(left=kill_MIRAMAR,right=agg_data, how='inner',left_on=['killer_name','match_id'], right_on=['player_name','match_id'])
```

```python
join_ERANGEL=pd.merge(left=kill_ERANGEL,right=agg_data, how='inner',left_on=['killer_name','match_id'], right_on=['player_name','match_id'])
```

```python
#중복되는 값 제거
join_MIRAMAR.drop_duplicates(["killer_name","match_id"])
```

```python
join_ERANGEL.drop_duplicates(["killer_name","match_id"])
```

```python
#게임 모드에 따라 데이터 나누기
solo_MIRAMAR=join_MIRAMAR[join_MIRAMAR['party_size']==1]
duo_MIRAMAR=join_MIRAMAR[join_MIRAMAR['party_size']==2]
squad_MIRAMAR=join_MIRAMAR[join_MIRAMAR['party_size']==4]

solo_ERANGEL=join_ERANGEL[join_ERANGEL['party_size']==1]
duo_ERANGEL=join_ERANGEL[join_ERANGEL['party_size']==2]
squad_ERANGEL=join_ERANGEL[join_ERANGEL['party_size']==4]
```

```python
#각 게임 모드마다 상위권, 하위권 나누기- 상위권: 20위 이내, 하위권 : 20~70등 내
#미라마 맵
solo_top_MIRAMAR=solo_MIRAMAR[solo_MIRAMAR['team_placement']<=20]
solo_low_MIRAMAR=solo_MIRAMAR[(solo_MIRAMAR['team_placement']>20)&(solo_MIRAMAR['team_placement']<=70)]

duo_top_MIRAMAR=duo_MIRAMAR[duo_MIRAMAR['team_placement']<=20]
duo_low_MIRAMAR=duo_MIRAMAR[(duo_MIRAMAR['team_placement']>20)&(duo_MIRAMAR['team_placement']<=70)]

squad_top_MIRAMAR=squad_MIRAMAR[squad_MIRAMAR['team_placement']<=20]
squad_low_MIRAMAR=squad_MIRAMAR[(squad_MIRAMAR['team_placement']>20)&(squad_MIRAMAR['team_placement']<=70)]

#에란겔 맵
solo_top_ERANGEL=solo_ERANGEL[solo_ERANGEL['team_placement']<=20]
solo_low_ERANGEL=solo_ERANGEL[(solo_ERANGEL['team_placement']>20)&(solo_ERANGEL['team_placement']<=70)]

duo_top_ERANGEL=duo_ERANGEL[duo_ERANGEL['team_placement']<=20]
duo_low_ERANGEL=duo_ERANGEL[(duo_ERANGEL['team_placement']>20)&(duo_ERANGEL['team_placement']<=70)]

squad_top_ERANGEL=squad_ERANGEL[squad_ERANGEL['team_placement']<=20]
squad_low_ERANGEL=squad_ERANGEL[(squad_ERANGEL['team_placement']>20)&(squad_ERANGEL['team_placement']<=70)]
```

```python
#상위권과 하위권 평균값 비교 데이터프레임 만들기
label=['player_assists','player_dbno','player_dist_ride', 'player_dist_walk','player_dmg', 'player_kills','player_survive_time']

def compare(df_top, df_low):
    temp_df=pd.DataFrame(columns=['top','low'],index=label)
    
    for i in range(len(temp_df)):
        temp_label=temp_df.index[i]
        temp_df.iloc[i,0]=df_top[temp_label].mean()
        temp_df.iloc[i,1]=df_low[temp_label].mean()
    return temp_df
```

```python
#미라마 비교
solo_compare_MIRAMAR=compare(solo_top_MIRAMAR,solo_low_MIRAMAR)
duo_compare_MIRAMAR=compare(duo_top_MIRAMAR,duo_low_MIRAMAR)
squad_compare_MIRAMAR=compare(squad_top_MIRAMAR,squad_low_MIRAMAR)
```

```python
#에란겔 비교
solo_compare_ERANGEL=compare(solo_top_ERANGEL,solo_low_ERANGEL)
duo_compare_ERANGEL=compare(duo_top_ERANGEL,duo_low_ERANGEL)
squad_compare_ERANGEL=compare(squad_top_ERANGEL,squad_low_ERANGEL)
```
