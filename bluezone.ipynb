```python
def count_death_time(df):
    df_minute=df.time//60
    death_count=pd.DataFrame(columns=['minutes','count'])
    for i in range(min(df_minute),max(df_minute)+1):
        temp_count=df_minute[df_minute.iloc[:]==i].count()
        death_count=death_count.append({"minutes":i,"count":temp_count},ignore_index=True)
    
    return death_count
```

```python
bluezone_count_ERANGEL=count_death_time(kill_bluezone_ERANGEL)
```

```python
bluezone_count_MIRAMAR=count_death_time(kill_bluezone_MIRAMAR)
```
